```
#converting color space
from skimage import color
from skimage import data
from pylab import *
#convertinh to hsv
img_hsv = color.rgb2hsv(img)
#converting back to rgb
img_rgb = color.hsv2rgb(img_hsv)
#show both the figures
figure(0)
io.imshow(img_hsv)
figure(1)
io.imshow(img_rgb)
```

```
#converting to xyz
img_xyz = color.rgb2xyz(img)
#converting back to rgb
img_rgbagain = color.xyz2rgb(img_xyz)
figure(0)
io.imshow(img_xyz)
figure(1)
io.imshow(img_rgbagain)
#converting rgb to lab
img_lab = color.rgb2lab(img)
figure(2)
io.imshow(img_lab)
#converting rgb to luv
img_luv = color.rgb2luv(img)
figure(3)
io.imshow(img_luv)
```

```
#converting rgb to yuv
img_yuv = color.rgb2yuv(img)
io.imshow(img_yuv)
```

```
#converting rgb to yiq
img_yiq = color.rgb2yiq(img)
io.imshow(img_yiq)
```

```
#converting rgb to lab
img_ypbpr = color.rgb2ypbpr(img)
io.imshow(img_ypbpr)
```

```
#resizing the image
from skimage.transform import resize
img_resized = resize(img,(300,300))
io.imshow(img_resized)
```

```
#rotating the image
from skimage.transform import rotate
img_rotate = rotate(img,angle = 45)
io.imshow(img_rotate)
```

```

img_rotated = rotate(img, angle=45, resize=True)
io.imshow(img_rotated)
```

```
from matplotlib import pyplot as plt

# selecting part of the image only 
cropped = img[100:(img.shape[0]-100),100:(img.shape[1]-100)]

plt.subplot(121), io.imshow(img)
plt.title('Original Image')
plt.subplot(122),io.imshow(cropped)
plt.title('Cropped Image')

plt.show()
```

```
cropped.size
```

```
#altering image exposure 
from skimage import exposure
image_bright = exposure.adjust_gamma(img, gamma=0.5,gain=1)
image_dark = exposure.adjust_gamma(img, gamma=1.5,gain=1)

# plotting images
plt.subplot(131), io.imshow(img)
plt.title('Original Image')

plt.subplot(132),io.imshow(image_bright)
plt.title('Bright Image')

plt.subplot(133),io.imshow(image_dark)
plt.title('Dark Image')

plt.show()
```

```
#draw lines
from skimage import draw
x,y = draw.line(0,0,200,200)
img[x,y] = 0
io.imshow(img)
```
